**Evaluate the Model**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

Here is a detailed explanation of the validation metrics imported from `sklearn.metrics`:

*   **Mean Squared Error (MSE):** Measures the average of the squares of the errors. It is the most common metric used for regression tasks. A lower MSE indicates a better fit of the model to the data.
    $$
    MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
    $$

*   **Mean Absolute Error (MAE):** Measures the average of the absolute differences between the actual and predicted values. It is less sensitive to outliers than MSE. A lower MAE indicates a better fit of the model to the data.
    $$
    MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
    $$

*   **R-squared (RÂ²):** Represents the proportion of the variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, where 1 indicates a perfect fit and 0 indicates that the model does not explain any of the variance.
    $$
    R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
    $$
    Where $SS_{res}$ is the sum of squares of the residuals and $SS_{tot}$ is the total sum of squares.

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

## Hyperparameter tuning

### Subtask:
Tune the hyperparameters of the Lasso model using techniques like GridSearchCV or RandomizedSearchCV to find the optimal `alpha` value.


**Reasoning**:
Tune the hyperparameters of the Lasso model using GridSearchCV to find the optimal alpha value by defining a parameter grid, instantiating the Lasso model and GridSearchCV, fitting GridSearchCV to the training data, and getting the best alpha.



```python
from sklearn.model_selection import GridSearchCV

param_grid = {'alpha': [0.1, 0.5, 1, 5, 10]}

lasso_model_tuned = Lasso(random_state=42)

grid_search = GridSearchCV(lasso_model_tuned, param_grid, cv=5)

grid_search.fit(X_train, y_train)

best_alpha = grid_search.best_params_['alpha']
print(f"Best alpha for Lasso: {best_alpha}")
```

## Re-evaluate lasso model

### Subtask:
Evaluate the performance of the tuned Lasso model using the selected metrics (MAE, MSE, RÂ²).


**Reasoning**:
Instantiate, train, and evaluate the tuned Lasso model and print the evaluation metrics.



```python
lasso_model_tuned = Lasso(alpha=best_alpha, random_state=42)
lasso_model_tuned.fit(X_train, y_train)
y_pred_lasso_tuned = lasso_model_tuned.predict(X_test)

mae_lasso_tuned = mean_absolute_error(y_test, y_pred_lasso_tuned)
mse_lasso_tuned = mean_squared_error(y_test, y_pred_lasso_tuned)
r2_lasso_tuned = r2_score(y_test, y_pred_lasso_tuned)

print(f"Tuned Lasso Model MAE: {mae_lasso_tuned}")
print(f"Tuned Lasso Model MSE: {mse_lasso_tuned}")
print(f"Tuned Lasso Model R2: {r2_lasso_tuned}")
```

## Compare tuned lasso model

### Subtask:

**Reasoning**:
Create a new row for the metrics_df DataFrame with the tuned Lasso model's performance metrics and append it to the DataFrame, then display the updated DataFrame.



```python
tuned_lasso_metrics = pd.DataFrame([{'Model': 'Tuned Lasso', 'MAE': mae_lasso_tuned, 'MSE': mse_lasso_tuned, 'RÂ²': r2_lasso_tuned}])
metrics_df = pd.concat([metrics_df, tuned_lasso_metrics], ignore_index=True)
display(metrics_df)
```
